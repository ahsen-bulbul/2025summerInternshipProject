```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# CSV dosyasını yükleme
df = pd.read_csv("/home/yapayzeka/ahsen_bulbul/data/aidevdb.yargitay_kararlar.csv")

# Histogram
sns.histplot(df['location'], bins=20)
plt.show()

# Boxplot
sns.boxplot(x=df['location'])
plt.show()

# Korelasyon matrisi
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()

```
